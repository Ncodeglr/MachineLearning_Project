#1.Load The Dataset

```python
import pandas as pd
train_pd = pd.read_csv("train.csv")
train_pd
```

```python
train_pd = train_pd[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
train_pd
```

```python
train_pd = train_pd.dropna()
train_pd
```

```python
train_pd.describe()
```

#2.Encoding & Preprocessing Data

```python
# to_categorical : Cette fonction convertit des étiquettes entières (ex: [0, 1, 2]) en vecteurs binaires "one-hot encoded" (ex: [[1,0,0], [0,1,0], [0,0,1]]).
# pclass-1 : to_categorical attend des étiquettes commencant à 0 (ex: 0 pour la 1ère classe, 1 pour la 2ème, etc.).
#            Ici, Pclass va de 1 à 3, donc on décale pour obtenir [0, 1, 2].

from tensorflow.keras.utils import to_categorical

pclass = train_pd['Pclass'].to_numpy()
as_categorical_pclass = to_categorical(pclass-1, num_classes = 3)
as_categorical_pclass
```

```python
train_pd['Sex'] = train_pd['Sex'].replace({'male' : 1, 'female': 0})
as_binary_sex = train_pd['Sex'].to_numpy()
as_binary_sex
```

```python
age_np = train_pd['Age'].to_numpy()
age_np = (age_np -age_np.mean()) / age_np.std()
age_np
```

```python
sibsp_np = train_pd['SibSp'].to_numpy()

#Scaling
sibsp_np= (sibsp_np -sibsp_np.mean()) / sibsp_np.std()
sibsp_np
```

```python
parch_np= train_pd['Parch'].to_numpy()

#Scaling
parch_np= (parch_np -parch_np.mean()) / parch_np.std()
parch_np
```

```python
fare = train_pd['Fare']
fare_np = fare.to_numpy()

#Scaling
fare_np = (fare_np -fare_np.mean()) / fare_np.std()
fare_np
```

```python
train_Y = train_pd['Survived'].to_numpy()
train_Y
```

#3. Model Implementation

```python
# np.concatenate : Assemble plusieurs tableaux le long d’un axe donné.
#         axis=0 : Empiler verticalement (ajouter des lignes).
#         axis=1 : Empiler horizontalement (ajouter des colonnes).

# reshape(-1, 1) convertit (712,) → (712, 1) pour chaque variable scalaire (non déjà encodée en one-hot).


from sklearn.linear_model import LogisticRegression
import numpy as np

train_X = np.concatenate((as_categorical_pclass,as_binary_sex.reshape(-1,1),
            age_np.reshape(-1,1),sibsp_np.reshape(-1,1),
            parch_np.reshape(-1,1),fare_np.reshape(-1,1)), axis=1)
train_X

#-------------------------------Logistic Regression Model-------------------------------------#

model = LogisticRegression(random_state=0).fit(train_X, train_Y)
model.score(train_X, train_Y)
```

# 4.Evaluation of the Machine Learning Model

```python
test_pd = pd.read_csv("test.csv")
test_pd = test_pd[['PassengerId','Pclass','Sex','Age','SibSp','Parch','Fare']]

#---------------------------Encoding & Preprocessing for the Dataset "test.csv"----------------------------#

test_pd = test_pd.fillna(test_pd['Age'].mean())

pclass = test_pd['Pclass'].to_numpy()
as_categorical_pclass = to_categorical(pclass-1, num_classes = 3)

test_pd['Sex'] = test_pd['Sex'].replace({'male' : 1, 'female': 0})
as_binary_sex = test_pd['Sex'].to_numpy()

age_np = test_pd['Age'].to_numpy()
age_np = (age_np - train_pd['Age'].mean()) / train_pd['Age'].std() #Scaling

sibsp_np = test_pd['SibSp'].to_numpy()
sibsp_np = (sibsp_np - train_pd['SibSp'].mean()) / train_pd['SibSp'].std() #Scaling

parch_np= test_pd['Parch'].to_numpy()
parch_np= (parch_np -train_pd['Parch'].mean()) / train_pd['Parch'].std() #Scaling

fare_np = test_pd['Fare'].to_numpy()
fare_np = (fare_np - train_pd['Fare'].mean()) / train_pd['Fare'].std() #Scaling


test_X = np.concatenate((as_categorical_pclass,as_binary_sex.reshape(-1,1),
            age_np.reshape(-1,1),sibsp_np.reshape(-1,1),
            parch_np.reshape(-1,1),fare_np.reshape(-1,1)), axis=1)
test_X
#-------------------------------------------------------------------------------------------------#
```

```python
predictions = model.predict(test_X)
passenger_ids = test_pd['PassengerId'].to_numpy()

sub_df = pd.DataFrame({'PassengerId' : passenger_ids, 'Survived' : predictions})
sub_df.to_csv('sub.csv', index=False)
```
