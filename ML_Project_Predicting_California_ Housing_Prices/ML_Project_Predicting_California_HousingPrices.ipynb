# 1.Load The Dataset

```python
import pandas as pd
housing_pd = pd.read_csv('housing.csv')
housing_pd.head()
```

```python
housing_pd['ocean_proximity'].value_counts()
```

```python
housing_pd_shuffled =housing_pd.sample(n=len(housing_pd), random_state=1)
housing_pd_shuffled
```

```python
pd.get_dummies(housing_pd_shuffled['ocean_proximity']).head()
```

```python
housing_pd_shuffled.drop('ocean_proximity',axis=1).head()
```

```python
housing_pd_final = pd.concat([housing_pd_shuffled.drop('ocean_proximity', axis=1),
                              pd.get_dummies(housing_pd_shuffled['ocean_proximity'])], axis=1)

housing_pd_final = housing_pd_final[['longitude',	'latitude',
                                     'housing_median_age',	'total_rooms',
                                     'total_bedrooms',	'population',
                                     'households',	'median_income',
                                     '<1H OCEAN',
                                     'INLAND',	'ISLAND',
                                     'NEAR BAY',	'NEAR OCEAN', 'median_house_value']]
housing_pd_final
```

```python
housing_pd_final = housing_pd_final.dropna()
len(housing_pd_final)
```

```python
housing_pd_final['<1H OCEAN'] = housing_pd_final['<1H OCEAN'].replace({True: 1, False: 0})
housing_pd_final['INLAND'] = housing_pd_final['INLAND'].replace({True: 1, False: 0})
housing_pd_final['ISLAND'] = housing_pd_final['ISLAND'].replace({True: 1, False: 0})
housing_pd_final['NEAR BAY'] = housing_pd_final['NEAR BAY'].replace({True: 1, False: 0})
housing_pd_final['NEAR OCEAN'] = housing_pd_final['NEAR OCEAN'].replace({True: 1, False: 0})
```

# 2.Slit into 3 Partitions: Train, Val and Test

```python
train_pd = housing_pd_final[:18000]
val_pd = housing_pd_final[18000:19217]
test_pd = housing_pd_final[19217:]
len(train_pd), len(val_pd), len(test_pd)
```

```python
X_train , Y_train = train_pd.to_numpy()[:,:-1] , train_pd.to_numpy()[:,-1]
X_val , Y_val = val_pd.to_numpy()[:,:-1] , val_pd.to_numpy()[:,-1]
X_test , Y_test = test_pd.to_numpy()[:,:-1] , test_pd.to_numpy()[:,-1]

(X_train.shape , Y_train.shape) , (X_val.shape , Y_val.shape) , (X_test.shape , Y_test.shape)
```

# 3.Preprocess Inputs

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler().fit(X_train[:,:8])
def preprocessor(X):
  A = np.copy(X)
  A[:,:8] = scaler.transform(A[:,:8])
  return A

X_train = preprocessor(X_train)
X_val = preprocessor(X_val)
X_test = preprocessor(X_test)

pd.DataFrame(X_train).head()
```

#4.Modele : Linear Regression


```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.linear_model import LinearRegression

lm = LinearRegression().fit(X_train, Y_train)

# Calcul du MSE
mse(lm.predict(X_train), Y_train), mse(lm.predict(X_val), Y_val)

# Calcul du RMSE
np.sqrt(mse(lm.predict(X_train), Y_train)), np.sqrt(mse(lm.predict(X_val), Y_val))
```
